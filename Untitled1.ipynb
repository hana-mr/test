```python
import wrds
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
MIN_YEAR = 1980
MAX_YEAR = 2003
```

```python
db = wrds.Connection()
print("Connexion WRDS établie. Début du téléchargement...")
```

```python
q_funda = f"""
    SELECT gvkey, datadate, fyear, at, dlc, dltt, oibdp, prcc_f, csho, pstkl, che, datafmt, consol
    FROM comp.funda
    WHERE datafmt='STD' AND consol='C' AND fyear BETWEEN {MIN_YEAR} AND {MAX_YEAR}
    """
compustat = db.raw_sql(q_funda, date_cols=['datadate'])
```

```python
print("Ajout des informations comp.company...")
q_company = """
    SELECT gvkey, sic, conm, loc
    FROM comp.company
"""
company = db.raw_sql(q_company)
compustat = compustat.merge(company, on='gvkey', how='left')
```

```python
db.close()
print("Téléchargement terminé et connexion fermée.")
```

```python
if not compustat.empty:
  compustat['sic'] = pd.to_numeric(compustat['sic'], errors='coerce')
  compustat = compustat[~compustat['sic'].between(6000, 6999)]
  compustat = compustat[~compustat['sic'].between(4900, 4999)]
  compustat = compustat.dropna(subset=['gvkey', 'fyear', 'at', 'prcc_f', 'csho'])
  compustat = compustat[compustat['at'] > 0]
  compustat['non_missing_count'] = compustat.notna().sum(axis=1)
  compustat = compustat.sort_values(['gvkey', 'fyear', 'non_missing_count'], ascending=[True, True, False])
  compustat = compustat.drop_duplicates(subset=['gvkey', 'fyear'], keep='first')
  compustat = compustat.drop(columns=['non_missing_count'])
  cols_to_winsor = ['dlc', 'dltt', 'at', 'oibdp', 'prcc_f', 'csho', 'pstkl', 'che']
  for col in cols_to_winsor:
    if col in compustat.columns:
      low, high = compustat[col].quantile([0.01, 0.99])
      compustat[col] = compustat[col].clip(lower=low, upper=high)
  compustat = compustat[compustat['loc'] == 'USA']
  print(f"Nettoyage terminé. Observations restantes: {len(compustat)}")
else:
  print("Nettoyage ignoré car le DataFrame est vide.")
```

```python
if not compustat.empty:
  compustat['total_debt'] = compustat['dlc'].fillna(0) + compustat['dltt'].fillna(0)
  compustat['market_equity'] = compustat['prcc_f'] * compustat['csho']
  compustat['book_leverage'] = compustat['total_debt'] / compustat['at']
  denom_market = compustat['total_debt'] + compustat['market_equity']
  compustat['market_leverage'] = compustat['total_debt'] / denom_market
  compustat['book_leverage'] = compustat['book_leverage'].clip(lower=0, upper=1)
  compustat['market_leverage'] = compustat['market_leverage'].clip(lower=0, upper=1)
  print("Variables de levier construites.")
```

```python
if not compustat.empty:
  firms_min_year = compustat[compustat['fyear'] == MIN_YEAR]['gvkey'].unique()
  firms_max_year = compustat[compustat['fyear'] == MAX_YEAR]['gvkey'].unique()
  survivor_gkeys = np.intersect1d(firms_min_year, firms_max_year)
  survivor_data = compustat[compustat['gvkey'].isin(survivor_gkeys)].copy()
  print(f"Nombre de firmes survivantes: {len(survivor_gkeys)}")
```

```python
summary_cols = ['book_leverage', 'market_leverage', 'at', 'total_debt']
index_names = {
'book_leverage': 'Book Leverage',
'market_leverage': 'Market Leverage',
'at': 'Total Assets (Mio USD)',
'total_debt': 'Total Debt (Mio USD)'}
```

```python
def generate_summary_table(df, title):
    summary = df[summary_cols].describe().T
    summary = summary.rename(columns={'mean': 'Mean', '50%': 'Median', 'std': 'Std. Dev.'})
    summary = summary.rename(index=index_names)
    summary['N'] = df.shape[0]
    print(f"\n--- {title} ---")
    print(tabulate(summary[['Mean', 'Median', 'Std. Dev.', 'N']],
                   headers='keys', tablefmt='github', floatfmt=".3f"))
    return summary

if not compustat.empty:
    summary_all = generate_summary_table(compustat, 'Tableau 1 - Panel A: All Firms (1980–2003)')
    summary_survivors = generate_summary_table(survivor_data, 'Tableau 1 - Panel B: Survivors (1980–2003)')
else:
    print
```

```python
if not compustat.empty:
    all_firms_avg = compustat.groupby('fyear', as_index=False)[['book_leverage', 'market_leverage']].mean()
    survivors_avg = survivor_data.groupby('fyear', as_index=False)[['book_leverage', 'market_leverage']].mean()

    plt.style.use('seaborn-v0_8-whitegrid')
    fig, axes = plt.subplots(2, 2, figsize=(14, 12))
    fig.suptitle('Figure 1: Capital Structure Ratios, 1980–2003 (Replication)', fontsize=16, y=1.02)

    axes[0, 0].plot(all_firms_avg['fyear'], all_firms_avg['book_leverage'], marker='o', markersize=4, color='darkblue', linewidth=2)
    axes[0, 0].set_title('Panel A: Book Leverage - All Firms', fontsize=14)
    axes[0, 0].set_ylabel('Average Book Leverage')

    axes[0, 1].plot(all_firms_avg['fyear'], all_firms_avg['market_leverage'], marker='o', markersize=4, color='darkred', linewidth=2)
    axes[0, 1].set_title('Panel B: Market Leverage - All Firms', fontsize=14)
    axes[0, 1].set_ylabel('Average Market Leverage')

    axes[1, 0].plot(survivors_avg['fyear'], survivors_avg['book_leverage'], marker='s', markersize=4, color='darkgreen', linewidth=2)
    axes[1, 0].set_title('Panel C: Book Leverage - Survivors', fontsize=14)
    axes[1, 0].set_xlabel('Fiscal Year')
    axes[1, 0].set_ylabel('Average Book Leverage')

    axes[1, 1].plot(survivors_avg['fyear'], survivors_avg['market_leverage'], marker='s', markersize=4, color='purple', linewidth=2)
    axes[1, 1].set_title('Panel D: Market Leverage - Survivors', fontsize=14)
    axes[1, 1].set_xlabel('Fiscal Year')
    axes[1, 1].set_ylabel('Average Market Leverage')

    for ax_row in axes:
        for ax in ax_row:
            ax.set_ylim(0.15, 0.45)
            ax.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.show()
    print("Figure 1 générée avec succès.")
else:
    print("Figure 1 ignorée car le DataFrame est vide.")
```
